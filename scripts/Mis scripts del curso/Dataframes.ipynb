# Manipulacion de Dataframes

```python
df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'values': range(6),
    'value2': np.random.randint(0,100,6)
})
```

```python
df
```

### Aggregate

```python
 df.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
df.groupby('key').aggregate({
    'values': 'min',
    'value2': 'max'
})
```

### Filter

Los filtros deben ser una funcion con una instruccion booleana 

```python
df.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std()>3
```

```python
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

### Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

### Apply

```python
def norm_by_col2(x):
    x['values'] /= x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

# Division de datos por columnas

```python
L = [0,1,0,1,2,0]
print(df)
df.groupby(L).sum()
```

`df.groupby(L).sum()` va a realizar una suma de los valores en el dataframe df segun array L. Es decir, segun el array L: `L = [0,1,0,1,2,0]`, va a sumar los valores del dataframe df, segun los valores que coincidan en las posiciones 0,1,2. Aqui, los valores 0 estan en las posiciones 0,2,5 por lo que sumará los valores que se encuentren en esas posiciones en df. De la isma manera con las posiciones 1 y 2.

```python
df.groupby('key').sum()
```

```python
df.groupby(df['key']).sum()
```

Esta linea es diferente de la anterior porque aqui estamos sumando segun los valores de esa clave, mientras que en la linea anterior estamos realizando sumas segun la clave. Al final dan el mismo resultado

```python
df2 = df.set_index('key')
df2
```

Modificamos el indice del dataframe. Antes el indice era 0,1,2,3,4,5. Ahora es A,B,C,A,B,C

```python
mapping = {
    'A': 'vowel',
    'B': 'consonant',
    'C': 'consonant'
}
```

```python
df2.groupby(mapping).sum()
```

Sumamos los valores que coincidian con la clave A y sumamos los valores que coincidian con la clave B y C porque realizamos una combinacion de ambas claves

```python
df2.groupby(str.lower).mean()
```

```python
df2.groupby([str.lower, mapping]).mean()
```

# Graficos desde dataframe y exportacion

```python
years = [year for year in range(1900, 2020)]
deaths = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    'year': years,
    'deaths': deaths
})
```

```python
df.head()
```

```python
df.plot(x = 'year', y = 'deaths', figsize = (10,8))
```

```python
df.plot(kind = "scatter", x = 'year', y = 'deaths', figsize = (10,8))
```

```python
my_plot = df.plot(x = 'year', y = 'deaths', figsize = (10,8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("../../../../../../Escritorio/dataframe_plot.png")
```
