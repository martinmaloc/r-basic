```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5,0.1)
x2 = np.arange(0,5,0.2)
x3 = np.arange(0,5,0.3)
x4 = np.arange(0,5,0.4)
```

`plt.subplot()` indica como estructurar la presentacion de varios graficos si usamos `plt.subplot(311)` estamos diciendo que estructure 3 filas y 1 fila donde en la primera fila coloque el grafico 1.

Y si usamos `plt.subplot(312)` estaremos respetando la estructura anterior pero ahora indicando que queremos graficar en la segunda fila al grafico 2

Y con `plt.subplot(313)` indicamos que la tercera fila coloque al 3er grafico

Por defecto `plt.subplot` esta en (111)

```python
plt.figure(1)
plt.subplot(221)
plt.plot(x1, f(x1), "ro", x2, f(x2), "k")

plt.subplot(222)
plt.plot(x2, f(x2), "g--")

plt.subplot(223)
plt.plot(x3, f(x3), "g:")

plt.subplot(224)
plt.plot(x4, f(x4), "g-")
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3,4])
plt.subplot(212)
plt.plot([4,5,6,7])

plt.figure(2)
plt.plot([1,2,3,4,5,6,7])

plt.figure(1)
plt.subplot(211)
plt.title("Esto es un titulo")
```

# Textos en los graficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, density = 1, facecolor = "g", alpha = 0.6)
plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$")
plt.text(120,0.016, r"$\mu = 100,\ \sigma = 20$")
plt.axis([20,180,0,0.025])
plt.grid(True)

plt.show()
```

`figsize =` es un parametro para definir el tamao de la figura en pulgadas y `dpi =` es un parametro para definir la cantidad de pixeles por pulgada 

```python
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(111)

x = np.arange(0, 10*np.pi, 0.1)
y = np.cos(x)

plt.plot(x,y, lw = 2)
plt.annotate(r"Mximo Local $4\pi$", xy = (4*np.pi, 1), xytext = (15, 1.5), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.annotate(r"Mnimo Local $3\pi$", xy = (3*np.pi, -1), xytext = (2.5, -1.5), 
             arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.xlim(0, 10*np.pi)
plt.ylim(-1.75,1.75)
plt.xlabel("X")
plt.ylabel("Coseno")
plt.show()
```

# Escalas 

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd * np.random.randn(1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize = (10,8))

plt.subplot(221)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(222)
plt.plot(x,y)
plt.yscale("log") # Es una escala semilogaritmica cuando unicamente uno de los dos ejes tiene escala logaritmica
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthresh = 0.01) # A que distancia queremos las divisiones logaritmicas. Aqui de 0.01 en 0.01
plt.title("Escala Log Simetrico")
plt.grid(True)

plt.subplot(224)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala Logistica")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.25, wspace = 0.25)

plt.show()
```

# Cambios en los ejes

`endpoint =` si indicamos True entonces siempre el ultimo digito va a ser el indicado en la funcion. Muchas veces el ultimo digito no es el especificado por lo que hay qu utilizar el parametro

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True) 
S, C = np.sin(x), np.cos(x)
```

`plt.xtiks` con esta funcion vamos a modificar los valores de el eje dado, en este casox. Entonces le indicamos que genere 9 numeros entre -4 y 4 y los coloque como corresponda en la escala del grafico

Lo mismo para `plt.yticks`

Lo que se hizo en `plt.ylim(x.min()*1.1, x.max()*1.1)` y `plt.ylim(S.min()*1.1, S.max()*1.1)` fue delimitar dinamicamente a los limites indicando que estos sean un 10% superiores al valor minimo y maximo de esos ejes

`plt.gca()` significa get current axis por lo que nos posibilita acceder a los ejes del grafico existente para modificarlos 

`plt.plot([x0,x0], [0,np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--")` aqui vamos a indicarle que grafique una linea entre las coordenadas $(X_0,\ 0)$ y $(X_0,\ \sin{x_0})$

`plt.annotate(r"$\sin(\frac{2\pi}{3} = \frac{\sqrt{3}}{2}$", xy = (x0, np.sin(x0)), xycoords = "data", xytext = (10,30), textcoords = "offset points") ` El parametro `xytext =` va a permitirnos indicar exactamente donde queremos que el texto de las coordenadas se ubique por lo que le damos dos coordenadas, pero al usar `textcoords = "offset points"` vamos a indicar que las coordenadas dadas en el parametro `xytext =` ya no representan coordenadas sino puntos hacia donde mover el texto. En este ejemplo le indicamos que mueva 10 puntos a la derecha  y 30 puntos hacia arriba

`for label in ax.get_xticklabels() + ax.get_yticklabels():` este bucle lo que esta haciendo es recorrer las etiquetas del eje x y del eje y a la vez por el simbolo +, y para cada una de ellas las guardo en labels

```python
plt.figure(figsize = (10,8))

#Graficamos Seno y Coseno, ademas de editar los ejes y sus valores
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "Seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "Coseno")
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$+\pi/2$", r"$+\pi$"])
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1", "", "+1"])

#Con las siguientes instrucciones vamos a modificar la posicion de los ejes como tambien eliminar los bordes de la figura
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data", 0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data", 0))

plt.legend(loc = "upper left")

#Agregamos texto y una linea a puntos especificos del grafico existente
x0 = 2 * np.pi/3
plt.plot([x0,x0], [0,np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue")
plt.annotate(r"$\sin\left(\frac{2\pi}{3}\right) = \frac{\sqrt{3}}{2}$", xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (20,10), textcoords = "offset points", fontsize = 16,
            arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3, rad = 0.2")) 

plt.plot([x0,x0], [0,np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
plt.annotate(r"$\cos\left(\frac{2\pi}{3}\right) = -\frac{1}{2}$", xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (30,10), textcoords = "offset points", fontsize = 16,
            arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3, rad = 0.2")) 

#Cambiamos el tamao de los ejes y agregamos una caja para superponer a las lineas
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(13)
    label.set_bbox(dict(facecolor = "white", edgecolor = "None", alpha = 0.6))

plt.show()
```
