# Matplotlib

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4]
```

```python
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de coordenadas")
plt.show() # Esta funcion une todas las intrucciones
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
```

```python
plt.plot(x,y, "ro")
plt.axis([0,6, 0,20])
plt.show() 
```

```python
import numpy as np
```

```python
data = np.arange(0.0, 10.0, 0.2)
data
```

```python
plt.plot(data, data, "r--", data, data**2, "bs", data, data**3, "g^")
plt.show()
```

```python
plt.plot(x,y, linewidth = 3)
```

```python
line, = plt.plot(x,y, "-")
line.set_antialiased(False)
```

```python
lines = plt.plot(data,data, data,data**2)
plt.setp(lines, color = "r", linewidth = 2) # Los parametros indicados afectan globalmente al grafico
```

```python
lines = plt.plot(data,data, data,data**2)
plt.setp(lines, "color", "r", "linewidth", 2)
```

```python
plt.plot(x,y, alpha = 0.5)
```

```python
plt.plot(x,y, marker = "+", linestyle = ":")
```
