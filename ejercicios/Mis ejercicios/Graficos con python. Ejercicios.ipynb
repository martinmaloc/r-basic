# Ejercicio de graficos con python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

### Ejercicio 1

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X) 
Y1 = Y + 1
Y2 = Y - 1

plt.figure(figsize = (10,8))

plt.plot(X, Y1, color='blue', alpha=1.00)
plt.plot(X, Y2, color='blue', alpha=1.00)

plt.fill_between(X, Y1, 1, color = "blue", alpha = 0.5)
plt.fill_between(X, Y2, -1, color = "blue", alpha = 0.5)
plt.fill_between(X, Y2, -1, where = (Y2 < -1), color = "pink")

plt.xlim(-np.pi, np.pi)
plt.xticks(())
plt.yticks(())
plt.show()
```

### Ejercicio 2

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n) 

plt.figure(figsize = (8,7))

plt.scatter(X,Y, c=np.arctan2(X, Y), cmap='rainbow', s=50, alpha=0.8)
plt.xlim(-3,3)
plt.ylim(-2,2)
plt.xticks(())
plt.yticks(())
plt.show()
```

### Ejercicio 3

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 

plt.figure(figsize = (10,8))
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 

for x, y in zip(X, Y1):     
        plt.text(x, y, '%.2f' % y, ha="center", va='bottom') 
for x, y in zip(X, -Y2):     
        plt.text(x, y - 0.08, '%.2f' % -y, ha="center", va='bottom')               
        
plt.ylim(-1.25,1.25)
plt.xticks(())
plt.yticks(())
plt.show()
```

### Ejercicio 4

```python
from matplotlib import cm
from matplotlib import colors
```

```python
Z = np.random.uniform(0, 1, 20)
Z = sorted(Z)
Z = np.array(Z)

explode_Z = []
for z in Z:
    if z == Z.max():
        explode_Z.append(0.15)
    else:
        explode_Z.append(0.05)
        
plt.figure(figsize = (10,8))

normdata = colors.Normalize(min(Z), max(Z))
colormap = cm.get_cmap("gray")
escalas_grises = colormap(normdata(Z))

centre_circle = plt.Circle((0,0), 0.05, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.pie(Z, explode = explode_Z, colors = escalas_grises, wedgeprops = dict(edgecolor = "black"))

plt.show()
```
